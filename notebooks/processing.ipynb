```python
import pandas as pd
df = pd.read_csv('retraction_05.txt', names=['data_id', 'accel0X', 'accel0Y', 'accel0Z', 'accel1X', 
                                             'accel1Y', 'accel1Z', 'tension', 'timestamp'])
df['time'] = pd.to_datetime(df['timestamp'], unit='ms')
df = df.drop(857635)
df
```

```python
from pandas.io.json import json_normalize
df_interface = pd.read_json('retraction_05.json', lines=True)
df_interface = json_normalize(df_interface.to_dict('records'))
```

```python
layer_ranges = df_interface[df_interface.status=='P'].groupby(
    'currentLayer').agg(['min', 'max'])[['timestamp']]
layer_ranges
```

```python
def get_layer_id(row):
    row_id =  layer_ranges[(layer_ranges['timestamp']['min'] <= row['time']) &
              (row['time'] <= layer_ranges['timestamp']['max'])
             ].head(1)
    try:
        return row_id.index[0]
    except:
        return -1

df['layer_id'] = df.apply(lambda x: get_layer_id(x), axis=1)
df
```

```python
df['accel0X'] = df['accel0X'].astype(float)
```

```python
df.plot('time', 'accel0X')
df.plot('time', 'accel0Y')
df.plot('time', 'accel0Z')
```

```python
x = df[(df['layer_id'] != -1)][['layer_id', 'accel0X']]
x['outlier'] = (x['accel0X'] < -1) | (x['accel0X'] > 1)
x = x.groupby('layer_id').sum().cumsum()
x.plot(y='outlier', use_index=True, marker=".")

y = df[(df['layer_id'] != -1)][['layer_id', 'accel0Y']]
y['outlier'] = (y['accel0Y'] < -2.0) | (y['accel0Y'] > 2.0)
y = y.groupby('layer_id').sum().cumsum()
y.plot(y='outlier', use_index=True, marker=".")

z = df[(df['layer_id'] != -1)][['layer_id', 'accel0Z']]
z['outlier'] = (z['accel0Z'] < 8.0) | (z['accel0Z'] > 12.0)
z = z.groupby('layer_id').sum().cumsum()
z.plot(y='outlier', use_index=True, marker=".")
```
